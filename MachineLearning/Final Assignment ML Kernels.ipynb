## Kernels: what are they?

Kernel methods are algorithms used to analyse patterns, and study general types of relations in datasets (classifications, clusters, principal components, and many others). 

Kernel methods come from kernel functions. **In machine learning, kernel functions are the ones known to perform the "kernel trick" - transforming linearly inseparable data into another dimension, to be able to divide or linearly separate the classes of data.**

Linearly separable data looks like the image below. How do we separate the blue and red data points in this case? It can easily be solved by drawing a line in between the two data classes. This "line" is called a separating hyperplane. We will talk about the hyperplane more in detail when we introduce the Support Vector Machine model later on.

<img src='linear.png'>

**But what do we do when our data is not linearly separable? No, we do not leave the problem unsolved! There is a specific tool that we saw called "kernel" that will help us with this.**

In the following image we can observe the problem visually. The dots are the data points in the feature space, representing different classes of data. What we see is that the data points are **not linearly separable**. We cannot just draw a straight line that separates the blue dots from the red ones. 

<img src='nonlinear.png'>

**So how can we separate these dots or classes of data points? This is where kernels come in handy.** We use them to transform linearly inseparable data into linearly separable data. The kernel function is applied to map the original non-linear observations into a higher dimensional space where they can be separated. This transformation is shown in the image below.

<img src='separable.png'>



**Our data points are linearly separable now! Thanks to the kernel trick.** We can now use a linear classifier in the new space of features. This the magic of applying kernels to solve the problem of non-linearly separable data points.



**So what exactly does the "kernel trick" do?** It is a function that computes  the inner product between the dots (or images) of all pairs of data in the feature space - *enabling the operation in a high-dimensional, implicit feature space, without actually having to compute the coordinates of the data in that space.* 
What that essentially means, is that kernels map the feature space and compute the inner dot product of the data pairs, thus making it work as if it was operating in an m dimension, without needing to actually go to that m dimensional space (where it is so costly to compute the data coordinates in). Kernels are a way of simplifying the computation steps that would normally be involved in the process. 

Feature mapping can be high-dimensional, which can be highly expensive computationally. The usual steps in feature mapping would consist of first calculating <f(x),f(y)> (*f is a mapping from n-dimension to m-dimension space*) and, thereafter, doing the dot product. Together, it would involve manipulations in the actual m dimensional space, being m a large number (resulting in a very high cost computationally). The kernel trick consists in replacing the dot products with an equivalent kernel function: 
> $k(x, x_0) = \Phi(x)T\Phi(x_0)$

With the kernel trick, the result obtained is the same, but the calculation is much easier, so it results in much lower computational costs. The kernel trick applies an inner dot product, making calculations much faster. Imagine applying this kernel trick to a large dataset, it would reduce the cost immensely.
**So, with kernels, we basically can work in infinite dimensions, without  having to go to the "m" dimensional space, that is, if we find the kernel that performs the calculation shortcut or "trick". **

**In simpler words, a kernel is a "shortcut" that enables one to do calculations much faster than if it required to make computations in a higher dimensional space.**

## Example of Cost Savings with Kernels

Here is an example that we found to be very illustrative to understand why kernels are used: Lili Jiang’s mathematical illustration. Retrieved from: https://www.quora.com/What-are-kernels-in-machine-learning-and-SVM-and-why-do-we-need-them/answer/Lili-Jiang?srid=oOgT

> **Simple Example.** We have "x" and "y", we already plugged in some numbers to make it more intuitive: 

    x = (1, 2, 3)
    y = (4, 5, 6)

>    The first step without using kernels would be to calculate f(x) and f(y)
    This would be the result:
    
    f(x) = (1, 2, 3, 2, 4, 6, 3, 6, 9)
    f(y) = (16, 20, 24, 20, 25, 30, 24, 30, 36)
    
>    The second step would require doing the dot product:

    <f(x), f(y)> 
    
>    This would be the end result:

    <f(x), f(y)> = 16 + 40 + 72 + 40 + 100+ 180 + 72 + 180 + 324 = 1024

>    This requires a lot of algebra, we go from mapping a 3-dimensional space to a 9-dimensional one. **Now let's try solving the same problem using the kernel function** (inner product of dots in the feature space):

    K(x, y) = (4 + 10 + 18 ) ^2 = 32^2 = 1024

**We get the same exact result with much less effort. The calculation is much less expensive computationally.**

## Kernels in Machine Learning

As previously stated, kernels are used to help us translate our data to a __plausibly infinite number of dimensions__ in order to find one where our data is __linearly separable__. With kernels, the “calculation is done for us” as it acceses an infinite number of dimensions, without having to visit that space, and return the solved problem. 

First, let's review the basics:
- How are kernels this implemented in Machine Learning? 
- What does SVM (Super Vector Machine) mean and its importance in the kernel environment?
- Why isn't Kernel a simple decision function?

Among the different tasks in Machine Learning, there is what we call: __Supervised Learning (SL)__. This is when, after training your model with the training data of your choice, you move on to predict parts of the data to which you already have an answer to, in order to measure the accuracy of your model's predictions.

__`SUPPORT VECTOR MACHINE (SVM)`__

 One of the most important models to use with kernels is the __Support Vector Machine or SVM__. SVMs are “supervised learning models with associated learning algorithms that analyze data for classification.” 
> __Source:__ Afonja, T. (2017). Kernel Functions – Towards Data Science. Retrieved from https://towardsdatascience.com/kernel-function-6f1d2be6091

SVMs are used to __analyze__ large datasets in order to identify __patterns__ from them. How do they do it? By generating partitions in data attributes, and then applying kernel functions to find the best possible hyperplane that divides data points into 2 categories. So what makes the best hyperplane? Well, it has to have a large __margin__ in order to separate the categories into different classes. This margin is represented as the __longest distance__ between the closest data points of those classes. The __optimal__ hyperplane is the one that has the largest distance to the nearest observation in both classes. All of this can be possible if observations are __linearly separable__.

> __" The larger the margin, the lower is the generalization error of the classifier "__

The wide use of SVM is due to its flexibility when training data because they do not consider __dependencies__ among attributes. 

__But what happens when the data is non-linearly separable?__

In order to solve this, we have to move to a space which has a higher dimension in order to achieve a linear separation. **This is where the kernel and SVM are best friends**: the SVM can use the help of the kernel function, to search for a higher dimension where the linearly inseparable data transfors to linearly separable data, finally obtaining a __scalar product__. 

The __advantages__ of using SVM are many. With the correct use of __kernel functions__, they can solve almost any classification problem, where there are different classes of data points to linearly separate. Furthermore, they are most precise when predicting, as they achieve the optimal hyperplane by the nearest points (known as support vectors) and not by the most distant point. This implies that SVMs are __robust and solid__ models.

On the other hand, they are __opaque__, as they have the risk of losing precision if the right kernel parameters aren't chosen. Another downside is the computation time which is usually very lenghty if it isn't correctly adjusted.


__`WHY KERNELS`__
-	Find a solution to classification problems, let us work in a higher dimension without increasing the cost.
-	Avoid high processing costs.

Kernels have become popular estimation and learning methods in ML, even with a greater mathematical application than traditional ML methods (neural networks).  
In the real world, non-linear methods are useful to detect dependencies, allowing to predict those properties of interest.
As long as we can formulate which kernel functions can be best for our model, we needn't to explicitly compute the coordinates of our data points into the high-dimensional feature space.  


> Common use of Kernels:
    >* __Polynomial Vector__
    >* __Radial Basis Function (RBF) Kernel__


Even though there are more (like Sigmoid) these ones are the solution to mostly all __classification problems__.

A computer program is said to learn from __experience E__ with respect to some __task T__ and some __performance measure P__, if its performance on T, as measured by P, improves with experience E.” 
* Tom Mitchell, Carnegie Mellon University

> __Example:__ Imagine you want your model to predict whether a specific recipe will return a "cupcake" or a "muffin" (task T). With machine learning algorithms and training of your model, it will learn from the data, and from the ingredient patterns (experience E), and in the case it has “learned” successfully , your model will then __do better at predicting which__ ingredient patterns return either a "cupcake" or a "muffin" (performance measure P).


__`Now… HOW ARE KERNELS APPLIED IN MACHINE LEARNING?`__

In SVM kernels are applied through a __similarity function__. When given two objects, the kernel outputs some similarity score. It is common to start from vectors with two values, but it can even be applied to integers and trees, anything that the kernel function knows how to compare with.
This is calculated with a __dot-product__. That is, given two vectors, the similarity here is the length of the projection of one vector onto another.
Another example, is the __Gaussian Kernel__. In which the similarity here, given two vectors, will be reduced with the use of the radius. This means that the distance between two objects now is re-weighted by the radius parameter.

* Kernels are used according to the specific problem that we are facing. 

__`A KERNEL ISNT A DECISION FUNCTION`: But why?__

Because it is used __inside__ the decision function. When confronted with a classification problem, this decision function uses the kernel to compare data to a number of support vectors that are weighted against the learning parameters (alpha) in an infinite number of dimensions. 

## Kernels in SVM

In this part, we will cover the practical side of kernels, commonly used with the Support Vector Machine model. We find the SVM model in Python's scikit learn library. But first, let's start by explaining what the SVM does.

A Support Vector Machine is a supervised machine learning algorithm which can be used for both classification and regression problems. It follows a technique called the kernel trick to transform the data and based on these transformations, it finds an optimal boundary between the possible outputs.

In simple words, it does some extremely complex data transformations to figure out how to separate the data based on the labels or outputs defined.

According to the Journal of Algorithms & Computational Technology (Vol. 8, No. 2):

> __Support Vector Machine:__ is a supervised statistical learning technique in the
field of machine learning. Based on the structural risk minimization, SVM tries
to find an optimized hyper-plane in a kernel space where training instances are
linearly separable. SVM is usually implemented through the so-called
soft-margin SVM because of its attractive characteristics listed as follows:
- (1) It possesses good generality under the principle of structural risk
minimization;
- (2) It can deal with non-linear problems by kernel methods;
- (3) It is robust to noisy instances after introducing slack variables;
- (4) It produces sparse solutions since the optimal hyper-plane depends only
on support vectors;
- (5) It guarantees convergence


>__Grid search__ is widely used in parameter selection or hyperparameter tuning due to its simplicity. It
conducts an exhaustive search on a finite set of parameters. The optimal
parameter corresponds to the one that has the highest score on a criterion. However, the success of grid
search stands on an assumption that the optimal value lies in the pre-defined
searching range; otherwise, grid search fails to work. Another problem is that
grid search is quite time-consuming especially when using classification
accuracy as the criterion. In machine learning, two tasks are commonly done at the same time in data pipelines: cross validation and (hyper)parameter tuning. Parameter tuning is the process to selecting the values for a model’s parameters that maximize the accuracy of the model.

> __cv__ stands for cross validation k-folds. We can specify the parameter cv = 5 for example, to indicate how many times it has to perform cross validation. Cross validation is the process of training learners using one set of data and testing it using a different set. 

>__svc_kernel__ is the kernel we want to try, and check if it is able to linearly separate the data points or support vectors. We can pass one or more than one option. With GridSearchCV, each kernel type will be tried to see which one can obtain a solution. Indicating ‘linear’ will use a linear hyperplane (a line in the case of 2D data) whereas ‘rbf’, 'sigmoid' and ‘poly’ use a non linear hyper-plane. We can also indicate ‘precomputed’, or a callable. The default value is ‘rbf’.

>__svc_C__ is the regularization parameter, commonly used to prevent overfitting, by assigning a softer or harder margin, i.e., accepting more errors if the margin is softer when c is low, or no errors if c is very high. This parameter controls the influence of each individual support vector; this process involves trading error penalty for stability.

>__svc_gamma__ is another parameter that we can specify on the SVM model, used for non linear hyperplanes (not necessary for 'linear' kernel). It is used as a similarity measure between two points. If gamma is large, the variance will be small, so the support vector will not have a wide-spread influence. Large gamma leads to low variance and high bias models, and vice-versa. The higher the gamma, the more it will try to exactly fit to the training dataset, leading to overfitting. A small gamma considers two points similar even if they are far from each other, a large gamma will consider two points similar _only_ if they are close to each other.

>__degree__ degree is a parameter used when kernel is set to ‘poly’, usually the degree is set at 3 by default. It is the degree of the polynomial used to find the hyperplane to split the data. Using degree = 1 is the same as using a ‘linear’ kernel. Also, increasing this parameters leads to higher training times. Other kernels will ignore the degree parameter, except 'poly'.

## Practical Examples of Kernels in SVM

### Linear SVM: Muffin vs. Cupcake

<img src='cupcake.jpeg'>

In this practical example of the application of kernels in SVM, we will load a database containing recipe information for muffins and cupcakes. Some people think muffins and cupcakes are the same thing, but in reality they have different proportions of ingredients in their recipes. In this case, we will use the SVM (supervised classification algorithm) to classify the recipe inputs into muffin (0) or cupcake (1). During the process, we will put kernels to practice, by seeing which one can divide the two recipes with a hyperplane.

```python
# Packages for analysis
import pandas as pd
import numpy as np
from sklearn import svm

# Packages for visuals
import matplotlib.pyplot as plt
import seaborn as sns; sns.set(font_scale=1.2)

# Allows charts to appear in the notebook
%matplotlib inline

# Pickle package
import pickle
```

```python
# Read in muffin and cupcake ingredient data
recipes = pd.read_csv('recipes_muffins_cupcakes.csv')
recipes.head(5)
```

The data is expressed as percentages of the total volume of each recipe. 

```python
# Plot two ingredients
sns.lmplot('Flour', 'Sugar', data=recipes, hue='Type',
           palette='Set1', fit_reg=False, scatter_kws={"s": 70});
```

We can see muffins and cupcakes are different, for example in the basic ingredientes such as flour and sugar, muffins have less sugar than cupcakes, but, on the other hand, they have much more flour.

```python
# Specify inputs for the model
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)

# ingredients = recipes[['Flour', 'Milk', 'Sugar', 'Butter', 'Egg', 'Baking Powder', 'Vanilla', 'Salt']].as_matrix()

ingredients = recipes[['Flour','Sugar']].as_matrix()
type_label = np.where(recipes['Type']=='Muffin', 0, 1) # muffin = 0; cupcake = 1 

# Feature names
recipe_features = recipes.columns.values[1:].tolist()
recipe_features
```

These variables above represent all the ingredients inside both cupcakes and muffins.

Now we will create our first SVM model, using the linear kernel because it is a classification problem that can be solved when the data is linearly separable. 

```python
# Fit the SVM model
model = svm.SVC(kernel= 'linear')
model.fit(ingredients, type_label)
```

```python
# Get the separating hyperplane
w = model.coef_[0]
a = -w[0] / w[1]
xx = np.linspace(30, 60)
yy = a * xx - (model.intercept_[0]) / w[1]

# Plot the parallels to the separating hyperplane that pass through the support vectors
b = model.support_vectors_[0]
yy_down = a * xx + (b[1] - a * b[0])
b = model.support_vectors_[-1]
yy_up = a * xx + (b[1] - a * b[0])
```

```python
# Plot the hyperplane
sns.lmplot('Flour', 'Sugar', data=recipes, hue='Type', palette='Set1', fit_reg=False, scatter_kws={"s": 70})
plt.plot(xx, yy, linewidth=2, color='black');
```

This is the optimal hyperplane obtained by the SVM model, using the linear kernel. The hyperplane divides the two types of recipes (muffin or cupcake).

```python
# Look at the margins and support vectors
sns.lmplot('Flour', 'Sugar', data=recipes, hue='Type', palette='Set1', fit_reg=False, scatter_kws={"s": 70})
plt.plot(xx, yy, linewidth=2, color='black')
plt.plot(xx, yy_down, 'k--')
plt.plot(xx, yy_up, 'k--')
plt.scatter(model.support_vectors_[:, 0], model.support_vectors_[:, 1],
            s=80, facecolors='none');
```

```python
# Create a function to guess when a recipe is a muffin or a cupcake
def muffin_or_cupcake(flour, sugar):
    if(model.predict([[flour, sugar]]))==0:
        print('You\'re looking at a muffin recipe!')
    else:
        print('You\'re looking at a cupcake recipe!')
```

```python
# Predict if 50 parts flour and 20 parts sugar
muffin_or_cupcake(50, 20)
```

```python
# Plot the point to visually see where the point lies
sns.lmplot('Flour', 'Sugar', data=recipes, hue='Type', palette='Set1', fit_reg=False, scatter_kws={"s": 70})
plt.plot(xx, yy, linewidth=2, color='black')
plt.plot(50, 20, 'yo', markersize='9');
```

```python
# Predict if 40 parts flour and 20 parts sugar
muffin_or_cupcake(40,20)
```

We will look at a facial recognition problem, to show how support vector machines work. The dataset used is the Wild dataset, which consists of several thousand photos of public figures, with labelled faces. A fetcher for the dataset is built into Scikit-Learn.


```python
from sklearn.datasets import fetch_lfw_people
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# use seaborn plotting defaults
import seaborn as sns; sns.set()


faces = fetch_lfw_people(min_faces_per_person=60)
print(faces.target_names)
print(faces.images.shape)
```

```python
fig, ax = plt.subplots(3, 5)
for i, axi in enumerate(ax.flat):
    axi.imshow(faces.images[i], cmap='bone')
    axi.set(xticks=[], yticks=[],
            xlabel=faces.target_names[faces.target[i]])
```

```python
from sklearn.svm import SVC
from sklearn.decomposition import PCA
from sklearn.pipeline import make_pipeline

pca = PCA(n_components=150, whiten=True, random_state=42)
svc = SVC(kernel = 'rbf', class_weight='balanced')
model = make_pipeline(pca, svc)
```

```python
from sklearn.model_selection import train_test_split
Xtrain, Xtest, ytrain, ytest = train_test_split(faces.data, faces.target, random_state=42)
```

```python
faces.data[1]
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import make_pipeline
from sklearn import preprocessing

param_grid = {'svc__kernel': ['rbf', 'sigmoid', 'poly'], # pass all parameters for GridSearchCV
              'svc__C': [1, 5, 10, 50],
              'svc__gamma': [0.0001, 0.0005, 0.001, 0.005]}


grid2 = GridSearchCV(model,
                     param_grid = param_grid, scoring = 'accuracy')

grid2.fit(Xtrain, ytrain)  #fitting data

print(grid2.best_params_)
```

```python
model = grid2.best_estimator_
yfit = model.predict(Xtest)
```

```python
fig, ax = plt.subplots(4, 6)
for i, axi in enumerate(ax.flat):
    axi.imshow(Xtest[i].reshape(62, 47), cmap='bone')
    axi.set(xticks=[], yticks=[])
    axi.set_ylabel(faces.target_names[yfit[i]].split()[-1],
                   color='black' if yfit[i] == ytest[i] else 'red')
fig.suptitle('Predicted Names; Incorrect Labels in Red', size=14);
```

```python
from sklearn.metrics import classification_report
print(classification_report(ytest, yfit,
                            target_names=faces.target_names))
```

**Our model has performed pretty well, with precision and recall of 86% and 85% respectively. The final and best parameters to obtain these metrics are: {'svc__C': 10, 'svc__gamma': 0.001, 'svc__kernel': 'rbf'}.**

# Summary & Conclusions

Once you reach the __linearly separable__ case by going through an infinite number of dimensions, you __optimize your model__ in order to get the highest accuracy. But sometimes classification problems arise due to a misclassification which leads to a low accuracy.
* This can be due to trained data which is insufficient to use to validate the model in order to obtain the best prediction.


It is important to emphasize, that __kernels cannot be applied__ if there isn’t at least two different classes. That is, Real problems often have more than 2 classes so it is important to Generalize the SVM to multi-class classification, for K > 2. 


On the other hand, in machine learning, the use of kernels as classifiers (as previously mentioned: SVM or KL, that is, kernel logistic regressor) are popular as they are known to achieve __state-of-the art__ performances for many classification problems. 

But even though most real-world data problems may not seem easily separable in linear terms, the use of kernels `isn’t efficient for all cases`, that is, it isn’t a solution for every classification case.  As training and prediction of learning algorithms could take up so much time and employ a high dimensional space by the use of the kernel trick.

For such, easily manageable data which can be easily assigned to a class, the __training of kernel classifiers__ is a total waste because fast and simple linear classifiers are enough.
 

In order to consider choosing one method or the other, two things have to be taken under consideration:

> 1. ___Kernel Mapping___: 
Mapping in SVM usually leads to the number of features blow up, that is, computing the mapping itself and even using the mapped representation (with the similarity function) can be counterproductive. So kernels are enough of a solution to overcome this problem. 
Thus, these explicit kernel mappings combine both the predictive power of nonlinear models with the scalability of linear models. Different from traditional dual kernels, explicit kernel methods can scale to millions of millions of samples. So in order to use these, we have to be careful with its parameters because they are __data dependent__ and can have an impact on the quality of the model as it is really sensitive. That is why, hyperparameters have to be tuned to obtained optimal values. All this is done implicitly in the model.
